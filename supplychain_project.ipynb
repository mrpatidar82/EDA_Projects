```python
#Kaunsa product sabse jyada revenue de raha hai.
revenue_product = df.groupby('Product type')['Revenue generated'].sum()
revenue_product.plot(kind='bar')
plt.title('Revenue by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Revenue')
plt.show()
```

```python
#Product Type Distribution
df['Product type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Product Type Distribution')
plt.ylabel('')
plt.show()
```

```python
#Supplier Name vs Revenue
supplier_rev = df.groupby('Supplier name')['Revenue generated'].sum()

supplier_rev.sort_values().plot(kind='barh')

plt.title('Revenue by Supplier')
plt.xlabel('Revenue')
plt.ylabel('Supplier')
plt.show()
```

```python
# Transportation Mode Usage
df['Transportation modes'].value_counts().plot(kind='bar')

plt.title('Transportation Mode Usage')
plt.xlabel('Transportation Mode')
plt.ylabel('Count')
plt.show()
```

```python
#Transportation Mode vs Shipping Cost
transport_cost = df.groupby('Transportation modes')['Shipping costs'].mean()

transport_cost.plot(kind='bar')

plt.title('Average Shipping Cost by Transportation Mode')
plt.xlabel('Transportation Mode')
plt.ylabel('Shipping Cost')
plt.show()
```

```python
#Product Type vs Defect Rates
defect_product = df.groupby('Product type')['Defect rates'].mean()

defect_product.plot(kind='bar')

plt.title('Defect Rate by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Defect Rate')
plt.show()
```

```python
#Revenue Distribution
plt.hist(df['Revenue generated'], bins=10)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
#Price vs Revenue Generated
#Price badhne se revenue badh raha hai ya nahi.
plt.scatter(df['Price'], df['Revenue generated'])
plt.title('Price vs Revenue')
plt.xlabel('Price')
plt.ylabel('Revenue')
plt.show()
```

```python
#Manufacturing Cost vs Revenue
#High manufacturing cost products jyada revenue dete hain ya nahi.
plt.scatter(df['Manufacturing costs'], df['Revenue generated'])
plt.title('Manufacturing Cost vs Revenue')
plt.xlabel('Manufacturing Cost')
plt.ylabel('Revenue')
plt.show()
```

```python
#Shipping Costs Distribution
plt.hist(df['Shipping costs'], bins=10)
plt.title('Shipping Cost Distribution')
plt.xlabel('Shipping Cost')
plt.ylabel('Frequency')
plt.show()
```

```python
#Stock Levels vs Products Sold
#High stock wale products jyada sell hue ya nahi.
plt.scatter(df['Stock levels'], df['Number of products sold'])

plt.title('Stock Levels vs Products Sold')
plt.xlabel('Stock Levels')
plt.ylabel('Products Sold')
plt.show()
```

```python
#Correlation Heatmap
#Sab numerical columns ke relations ek sath.
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Lead Time vs Shipping Time
#Lead time aur shipping time ka relation.
plt.scatter(df['Lead times'], df['Shipping times'])
plt.title('Lead Time vs Shipping Time')
plt.xlabel('Lead Time')
plt.ylabel('Shipping Time')
plt.show()
```

```python
#Inspection Results Count
#Pass / Fail / Pending count.
df['Inspection results'].value_counts().plot(kind='bar')
plt.title('Inspection Results')
plt.xlabel('Inspection Result')
plt.ylabel('Count')
plt.show()
```

```python
#Boxplot for Outliers
#Price Outliers
plt.boxplot(df['Price'])

plt.title('Price Boxplot')
plt.show()
```
