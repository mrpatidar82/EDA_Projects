```python
# 3. TOP 10 COUNTRIES
top_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))

top_country.sort_values().plot(kind='barh')

plt.title("Top 10 Countries")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
# 4. TOP 10 DIRECTORS
top_director = df['director'].value_counts().head(10)

plt.figure(figsize=(10,6))

top_director.sort_values().plot(kind='barh')

plt.title("Top 10 Directors")
plt.xlabel("Count")
plt.ylabel("Director")
plt.show()
```

```python
# 5. RATING DISTRIBUTION
plt.figure(figsize=(10,6))

df['rating'].value_counts().plot(kind='bar')

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. TOP GENRES
genre = df['listed_in'].str.split(', ', expand=True).stack()

top_genre = genre.value_counts().head(10)

plt.figure(figsize=(10,6))

top_genre.sort_values().plot(kind='barh')

plt.title("Top Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
#7. MOVIE DURATION ANALYSIS
movies = df[df['type'] == 'Movie'].copy()

movies['duration_int'] = movies['duration'] \
    .str.replace(' min', '', regex=False)

movies['duration_int'] = pd.to_numeric(
    movies['duration_int'],
    errors='coerce'
)

movies = movies.dropna(
    subset=['duration_int']
)

plt.figure(figsize=(10,6))

plt.hist(
    movies['duration_int'],
    bins=20
)

plt.title("Movie Duration Distribution")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Count")

plt.show()
```

```python
# QUESTION 8
# TV SHOW SEASONS ANALYSIS
# ==========================================

tv = df[df['type'] == 'TV Show'].copy()

tv['season_count'] = tv['duration'] \
    .str.replace(
        ' Seasons',
        '',
        regex=False
    )

tv['season_count'] = tv['season_count'] \
    .str.replace(
        ' Season',
        '',
        regex=False
    )

tv['season_count'] = pd.to_numeric(
    tv['season_count'],
    errors='coerce'
)

tv = tv.dropna(
    subset=['season_count']
)

season_counts = tv['season_count'] \
    .value_counts() \
    .sort_index()

plt.figure(figsize=(10,6))

plt.bar(
    season_counts.index.astype(str),
    season_counts.values
)

plt.title("TV Show Seasons Distribution")
plt.xlabel("Number of Seasons")
plt.ylabel("Count")

plt.show()
```

```python
# 9. TYPE VS RATING
type_rating = pd.crosstab(df['rating'], df['type'])

type_rating.plot(kind='bar', stacked=True, figsize=(12,6))

plt.title("Type vs Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. COUNTRY VS TYPE
top_countries = df['country'].value_counts().head(10).index

country_type = df[df['country'].isin(top_countries)]

ct = pd.crosstab(country_type['country'],
                 country_type['type'])

ct.plot(kind='bar', figsize=(12,6))

plt.title("Country vs Type")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. GENRE VS TYPE
genre_type = pd.crosstab(df['listed_in'], df['type'])

genre_type.head(10).plot(kind='bar',
                         figsize=(12,6))

plt.title("Genre vs Type")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
## QUESTION 12
# CONTENT ADDED PER YEAR
# ==========================================

df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

df['year_added'] = df[
    'date_added'
].dt.year

year_added = df['year_added'] \
    .value_counts() \
    .sort_index()

plt.figure(figsize=(12,6))

plt.plot(
    year_added.index,
    year_added.values
)

plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
# 13. TOP ACTORS
actors = df['cast'].str.split(', ', expand=True).stack()

top_actors = actors.value_counts().head(10)

plt.figure(figsize=(10,6))

top_actors.sort_values().plot(kind='barh')

plt.title("Top Actors")
plt.xlabel("Count")
plt.ylabel("Actor")
plt.show()

# ==============================
# END OF EDA
# ==============================

print("EDA COMPLETED SUCCESSFULLY")
```

```python
df['year_added'] = df['date_added'].dt.year

df['month_added'] = df['date_added'].dt.month

df['day_added'] = df['date_added'].dt.day
```

```python
# 2. MOVIES VS TV SHOWS OVER TIME
type_year = pd.crosstab(
    df['year_added'],
    df['type']
)

plt.figure(figsize=(12,6))

plt.plot(
    type_year.index,
    type_year['Movie'],
    label='Movies'
)

plt.plot(
    type_year.index,
    type_year['TV Show'],
    label='TV Shows'
)

plt.legend()

plt.title("Movies vs TV Shows Over Time")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
# 3. MONTHLY CONTENT ADDITION
monthly = df['month_added'] \
    .value_counts() \
    .sort_index()

plt.figure(figsize=(10,5))

plt.plot(
    monthly.index,
    monthly.values,
    marker='o'
)

plt.title("Monthly Content Addition")
plt.xlabel("Month")
plt.ylabel("Count")

plt.show()
```

```python
# 4. COUNTRY VS TYPE HEATMAP
top_country = df['country'] \
    .value_counts() \
    .head(10) \
    .index

country_df = df[
    df['country'].isin(top_country)
]

country_type = pd.crosstab(
    country_df['country'],
    country_df['type']
)

plt.figure(figsize=(10,6))

sns.heatmap(
    country_type,
    annot=True,
    fmt='d'
)

plt.title("Country vs Type Heatmap")

plt.show()
```

```python
# 5. RATING VS TYPE HEATMAP
rating_type = pd.crosstab(
    df['rating'],
    df['type']
)

plt.figure(figsize=(10,6))

sns.heatmap(
    rating_type,
    annot=True,
    fmt='d'
)

plt.title("Rating vs Type Heatmap")

plt.show()
```

```python
# 6. TOP LONGEST MOVIES
movies = df[df['type'] == 'Movie'].copy()

movies['duration_int'] = movies['duration'] \
    .str.replace(' min', '', regex=False)

movies['duration_int'] = pd.to_numeric(
    movies['duration_int'],
    errors='coerce'
)

movies = movies.dropna(
    subset=['duration_int']
)

longest_movies = movies[
    ['title', 'duration_int']
].sort_values(
    by='duration_int',
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

plt.barh(
    longest_movies['title'],
    longest_movies['duration_int']
)

plt.title("Top Longest Movies")
plt.xlabel("Duration")
plt.ylabel("Movie")

plt.show()
```

```python
# 7. GENRE ANALYSIS HEATMAP
genre_df = df.copy()

genre_df['listed_in'] = genre_df[
    'listed_in'
].str.split(', ')

genre_df = genre_df.explode(
    'listed_in'
)

genre_type = pd.crosstab(
    genre_df['listed_in'],
    genre_df['type']
)

genre_type = genre_type.head(15)

plt.figure(figsize=(10,8))

sns.heatmap(
    genre_type,
    annot=True,
    fmt='d'
)

plt.title("Genre vs Type Heatmap")

plt.show()
```

```python
# 8. COUNTRY WISE CONTENT
country_content = df['country'] \
    .value_counts() \
    .head(15)

plt.figure(figsize=(12,6))

country_content.plot(
    kind='bar'
)

plt.title("Country Wise Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
# 9. MOST COMMON RATINGS
rating_count = df['rating'] \
    .value_counts()

plt.figure(figsize=(10,6))

plt.pie(
    rating_count.values,
    labels=rating_count.index,
    autopct='%1.1f%%'
)

plt.title("Content Rating Distribution")

plt.show()
```
